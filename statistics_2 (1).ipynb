```python
1.  What is a random variable in probability theory?
-   Random Variable: A numerical value representing the outcomes of a random experiment.
2.  What are the types of random variables?
Types:
  Discrete Random Variables: Take on a countable number of distinct values (e.g., number of heads in 10 coin flips).
  Continuous Random Variables: Take on an infinite number of possible values within a given range (e.g., height of students).
3.  What is the difference between discrete and continuous distributions?
-   Discrete vs. Continuous Distributions
  Discrete Distributions: Deal with discrete random variables (e.g., binomial distribution).
  Continuous Distributions: Deal with continuous random variables (e.g., normal distribution).
4. What are probability distribution functions (PDF)?
-  A PDF describes the likelihood of different outcomes for a random variable. It gives the probability that a random variable takes on a specific value (discrete) or falls within a range of values (continuous).
5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
-  A CDF gives the probability that a random variable is less than or equal to a specific value. It accumulates probabilities up to that value.
* Difference from PDF: PDF provides probabilities for specific values/ranges; CDF provides cumulative probabilities up to a certain value.
6. What is a discrete uniform distribution?
-  A discrete uniform distribution assigns equal probability to each possible outcome in a finite set of equally likely values (e.g., rolling a fair die).
7. What are the key properties of a Bernoulli distribution?
-  Bernoulli Distribution
Key Properties:
    Two possible outcomes: success (1) and failure (0).
    Probability of success = 𝑝; probability of failure = 1−𝑝.
8. What is the binomial distribution, and how is it used in probability?
-  Models the number of successes in a fixed number of independent trials of a binary experiment.
Use: To calculate probabilities of a given number of successes in a series of trials (e.g., flipping a coin 10 times).
9. What is the Poisson distribution and where is it applied?
-  Models the number of events occurring within a fixed interval of time or space.
Applications: Customer arrivals, phone call frequencies, etc.
10. What is a continuous uniform distribution?
-   A continuous uniform distribution assigns equal probability to all values within a specified interval.
11. What are the characteristics of a normal distribution?
-   Characteristics:
  Symmetric, bell-shaped curve.
  Mean, median, and mode are equal.
  Defined by its mean (μ) and standard deviation (σ).
12. What is the standard normal distribution, and why is it important?
-   A normal distribution with a mean of 0 and a standard deviation of 1. It is important for standardizing data and comparing different distributions.
13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
-   The CLT states that the distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the original population's distribution.
    Importance: Allows the use of normal distribution for inference in various scenarios.
14. How does the Central Limit Theorem relate to the normal distribution?
-   The CLT explains why the normal distribution is ubiquitous in statistics and why sample means tend to be normally distributed.
15. What is the application of Z statistics in hypothesis testing?
-   Z-statistics are used to determine if a sample mean significantly differs from a population mean, aiding in hypothesis testing.
16. How do you calculate a Z-score, and what does it represent?
-   Calculating a Z-Score
    𝑍=(𝑋−𝜇)/𝜎
Represents: The number of standard deviations a data point is from the mean.
17. What are point estimates and interval estimates in statistics?
-   Point Estimates: Single values (e.g., sample mean) used to estimate population parameters.
    Interval Estimates: Range of values (e.g., confidence intervals) that likely contain the population parameter.
18. What is the significance of confidence intervals in statistical analysis?
-   Confidence intervals provide a range of plausible values for population parameters, offering more information than point estimates.
19. What is the relationship between a Z-score and a confidence interval?
-   Z-scores are used to calculate confidence intervals by determining the number of standard deviations a sample mean is from the population mean.
20. How are Z-scores used to compare different distributions?
-   Z-scores standardize data, allowing comparisons across different distributions.
21. What are the assumptions for applying the Central Limit Theorem?
-   Independent observations.
    Random sampling.
    Sample size sufficiently large (n > 30 is often used as a rule of thumb).
22. What is the concept of expected value in a probability distribution?
-   The expected value is the long-term average or mean value of a random variable. It is calculated by multiplying each possible value by its probability and summing the results.
23. How does a probability distribution relate to the expected outcome of a random variable?
-   The probability distribution defines the expected outcomes of a random variable by specifying the probabilities associated with each value.
# PRACTICAL
1. Write a Python program to generate a random variable and display its value.
-  import random
random_variable = random.random()
print("Random Variable:", random_variable)
2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
-  import numpy as np
   import matplotlib.pyplot as plt

values = np.arange(1, 7)
probabilities = np.ones(6) / 6

plt.stem(values, probabilities, use_line_collection=True)
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Discrete Uniform Distribution PMF')
plt.show()

3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
-   def bernoulli_pdf(p, x):
    if x in [0, 1]:
        return p if x == 1 else (1 - p)
    else:
        return 0
p = 0.5
x = 1
print("Bernoulli PDF:", bernoulli_pdf(p, x))

4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.
-  n = 10
p = 0.5
size = 1000

# Simulate binomial distribution
binomial_data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(binomial_data, bins=n+1, edgecolor='black', density=True)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution Histogram')
plt.show()

5. Create a Poisson distribution and visualize it using Python.
-  lambda_ = 3
size = 1000

# Create Poisson distribution
poisson_data = np.random.poisson(lambda_, size)

# Plot histogram
plt.hist(poisson_data, bins=range(0, max(poisson_data)+1), edgecolor='black', density=True)
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution')
plt.show()

6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution.
- values = np.arange(1, 7)
probabilities = np.ones(6) / 6
cdf = np.cumsum(probabilities)

# Plot CDF
plt.step(values, cdf, where='mid')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('Discrete Uniform Distribution CDF')
plt.show()

7. Generate a continuous uniform distribution using NumPy and visualize it.
-  # Generate continuous uniform distribution
continuous_uniform_data = np.random.uniform(0, 1, 1000)

# Plot histogram
plt.hist(continuous_uniform_data, bins=20, edgecolor='black', density=True)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Continuous Uniform Distribution')
plt.show()

8. Simulate data from a normal distribution and plot its histogram.
-  mean = 0
std_dev = 1
size = 1000

# Simulate normal distribution
normal_data = np.random.normal(mean, std_dev, size)

# Plot histogram
plt.hist(normal_data, bins=30, edgecolor='black', density=True)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution')
plt.show()

9. Write a Python function to calculate Z-scores from a dataset and plot them.
-  def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

# Example dataset
data = np.random.normal(0, 1, 100)

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot Z-scores
plt.hist(z_scores, bins=30, edgecolor='black', density=True)
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.title('Z-scores')
plt.show()

10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
-   # Non-normal distribution (exponential distribution)
exponential_data = np.random.exponential(scale=2, size=1000)

# Sample means
sample_means = []
sample_size = 30
num_samples = 1000

for _ in range(num_samples):
    sample = np.random.choice(exponential_data, sample_size)
    sample_means.append(np.mean(sample))

# Plot histogram of sample means
plt.hist(sample_means, bins=30, edgecolor='black', density=True)
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title('Central Limit Theorem (CLT)')
plt.show()


11. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.
-   import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mean = 50
std_dev = 10
sample_size = 30
num_samples = 1000

# Generate multiple samples and calculate sample means
sample_means = [np.mean(np.random.normal(mean, std_dev, sample_size)) for _ in range(num_samples)]

# Plot histogram of sample means
plt.hist(sample_means, bins=30, edgecolor='black', density=True)
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.title('Central Limit Theorem')
plt.show()

12. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)5
-   def plot_standard_normal_distribution():
    import numpy as np
    import matplotlib.pyplot as plt

    # Generate standard normal distribution data
    x = np.linspace(-4, 4, 1000)
    y = (1 / (np.sqrt(2 * np.pi))) * np.exp(-0.5 * x**2)

    # Plot the standard normal distribution
    plt.plot(x, y)
    plt.xlabel('Z')
    plt.ylabel('Probability Density')
    plt.title('Standard Normal Distribution')
    plt.grid(True)
    plt.show()

# Call the function to plot
plot_standard_normal_distribution()

13. Generate random variables and calculate their corresponding probabilities using the binomial distribution.
-   from scipy.stats import binom

# Parameters for the binomial distribution
n = 10
p = 0.5
k = np.arange(0, n+1)

# Calculate the probabilities
binomial_probs = binom.pmf(k, n, p)

# Print probabilities
print("Random Variables:", k)
print("Corresponding Probabilities:", binomial_probs)

14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution.
-   def calculate_z_score(data_point, mean, std_dev):
    return (data_point - mean) / std_dev

# Example data point
data_point = 75
mean = 70
std_dev = 10

# Calculate Z-score
z_score = calculate_z_score(data_point, mean, std_dev)
print("Z-score:", z_score)

15. Implement hypothesis testing using Z-statistics for a sample dataset.
-   from scipy.stats import norm

# Example dataset
sample_mean = 75
population_mean = 70
std_dev = 10
sample_size = 30

# Calculate Z-statistic
z_statistic = (sample_mean - population_mean) / (std_dev / np.sqrt(sample_size))
print("Z-statistic:", z_statistic)

# Calculate p-value
p_value = 2 * (1 - norm.cdf(np.abs(z_statistic)))
print("p-value:", p_value)

16. Create a confidence interval for a dataset using Python and interpret the result
-   import scipy.stats as stats

# Example dataset
data = [75, 80, 85, 90, 95, 100]

# Calculate sample mean and standard error
mean = np.mean(data)
std_err = stats.sem(data)

# Calculate confidence interval (95%)
conf_interval = stats.t.interval(0.95, len(data)-1, loc=mean, scale=std_err)
print("Confidence Interval:", conf_interval)

17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
-   data = np.random.normal(70, 10, 100)

# Calculate sample mean and standard error
mean = np.mean(data)
std_err = stats.sem(data)

# Calculate confidence interval (95%)
conf_interval = stats.t.interval(0.95, len(data)-1, loc=mean, scale=std_err)
print("Confidence Interval:", conf_interval)

18. Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution
-   from scipy.stats import norm

# Generate data for PDF
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

# Plot PDF
plt.plot(x, y)
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution PDF')
plt.grid(True)
plt.show()

19. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution
-   from scipy.stats import poisson

lambda_ = 3
x = np.arange(0, 15)

# Calculate CDF
poisson_cdf = poisson.cdf(x, lambda_)

# Plot CDF
plt.step(x, poisson_cdf, where='mid')
plt.xlabel('Number of Events')
plt.ylabel('Cumulative Probability')
plt.title('Poisson Distribution CDF')
plt.grid(True)
plt.show()

20. Simulate a random variable using a continuous uniform distribution and calculate its expected value
-   # Generate continuous uniform distribution data
data = np.random.uniform(0, 1, 1000)

# Calculate expected value
expected_value = np.mean(data)
print("Expected Value:", expected_value)

21. Write a Python program to compare the standard deviations of two datasets and visualize the difference
-   data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 15, 100)

std_dev1 = np.std(data1)
std_dev2 = np.std(data2)

print("Standard Deviation of Dataset 1:", std_dev1)
print("Standard Deviation of Dataset 2:", std_dev2)

# Plot histograms
plt.hist(data1, bins=30, alpha=0.5, label='Dataset 1')
plt.hist(data2, bins=30, alpha=0.5, label='Dataset 2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Comparison of Standard Deviations')
plt.legend()
plt.show()

22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
-   data = np.random.normal(70, 10, 100)

# Calculate range
data_range = np.ptp(data)
print("Range:", data_range)

# Calculate IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
print("IQR:", iqr)

23. Implement Z-score normalization on a dataset and visualize its transformation
-   data = np.random.normal(70, 10, 100)

# Calculate Z-scores
mean = np.mean(data)
std_dev = np.std(data)
z_scores = (data - mean) / std_dev

# Plot original data and Z-scores
plt.hist(data, bins=30, alpha=0.5, label='Original Data')
plt.hist(z_scores, bins=30, alpha=0.5, label='Z-scores')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Z-score Normalization')
plt.legend()
plt.show()

24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution
-   from scipy.stats import skew, kurtosis

data = np.random.normal(70, 10, 100)

# Calculate skewness and kurtosis
skewness = skew(data)
kurtosis_value = kurtosis(data)

print("Skewness:", skewness)
print("Kurtosis:", kurtosis_value)
```
